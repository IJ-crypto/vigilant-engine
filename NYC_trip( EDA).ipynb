```python
train=df_train
test=df_test
```

```python
train['pickup_date']=train['pickup_datetime'].apply(lambda x:x.split(' ')[0])
train['pickup_time']=train['pickup_datetime'].apply(lambda x:x.split(' ')[1])
test['pickup_date']=test['pickup_datetime'].apply(lambda x:x.split(' ')[0])
test['pickup_time']=test['pickup_datetime'].apply(lambda x:x.split(' ')[1])
train['dropoff_date']=train['dropoff_datetime'].apply(lambda x:x.split(' ')[0])
train['dropoff_time']=train['dropoff_datetime'].apply(lambda x:x.split(' ')[1])
train.head()
```

```python
train.drop('pickup_datetime',axis=1,inplace=True)
train.drop('dropoff_datetime',axis=1,inplace=True)
test.drop('pickup_datetime',axis=1,inplace=True)
```

```python
train['pickup_year']=train['pickup_date'].apply(lambda x:x.split('-')[0]).astype('int')
train['pickup_month']=train['pickup_date'].apply(lambda x:x.split('-')[1]).astype('int')
train['pickup_day']=train['pickup_date'].apply(lambda x:x.split('-')[2]).astype('int')

test['pickup_year']=test['pickup_date'].apply(lambda x:x.split('-')[0]).astype('int')
test['pickup_month']=test['pickup_date'].apply(lambda x:x.split('-')[1]).astype('int')
test['pickup_day']=test['pickup_date'].apply(lambda x:x.split('-')[2]).astype('int')

train['dropoff_year']=train['dropoff_date'].apply(lambda x:x.split('-')[0]).astype('int')
train['dropoff_month']=train['dropoff_date'].apply(lambda x:x.split('-')[1]).astype('int')
train['dropoff_day']=train['dropoff_date'].apply(lambda x:x.split('-')[2]).astype('int')

train_pickup_hr=train['pickup_time'].apply(lambda x:x.split(':')[0]).astype('int')
train_pickup_min=train['pickup_time'].apply(lambda x:x.split(':')[1]).astype('int')
train_pickup_sec=train['pickup_time'].apply(lambda x:x.split(':')[2]).astype('int')

test_pickup_hr=test['pickup_time'].apply(lambda x:x.split(':')[0]).astype('int')
test_pickup_min=test['pickup_time'].apply(lambda x:x.split(':')[1]).astype('int')
test_pickup_sec=test['pickup_time'].apply(lambda x:x.split(':')[2]).astype('int')

dropoff_hr=train['dropoff_time'].apply(lambda x:x.split(':')[0]).astype('int')
dropoff_min=train['dropoff_time'].apply(lambda x:x.split(':')[1]).astype('int')
dropoff_sec=train['dropoff_time'].apply(lambda x:x.split(':')[2]).astype('int')

train['pickup_time']= 3600*train_pickup_hr + 60*train_pickup_min + train_pickup_sec
test['pickup_time']= 3600*test_pickup_hr + 60*test_pickup_min + test_pickup_sec
train['dropoff_time']= 3600*dropoff_hr + dropoff_min*60 + dropoff_sec                                                                            
                                                                                  
train.drop(['pickup_date','dropoff_date'],axis=1,inplace=True)
test.drop(['pickup_date'],axis=1,inplace=True)
 
train.head()                                                                         
```
