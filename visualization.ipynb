```python
import pandas as np
import numpy as np
import matplotlib.pyplot as plt
from io import StringIO, BytesIO
```

```python
x = np.arange(0,10)
y = np.arange(21,31)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('2D Graph')
plt.savefig('graph.png')
```

```python
plt.plot(x,y)
```

```python
y= x*x
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'yo-')
```

```python
x= np.arange(0,4*np.pi,0.2)
y_s = np.sin(x)
y_c = np.cos(x)
plt.title('Sine Wave')
plt.xlabel('Values')
plt.ylabel('Sin(Values)')
plt.plot(x,y)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_s,'g--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_c)
plt.title('Cosine')
plt.show()
```
